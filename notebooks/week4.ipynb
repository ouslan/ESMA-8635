```python
import pymc as pm
import pandas as pd
import arviz as az
import bambi as bmb

import matplotlib.pyplot as plt

az.style.use("arviz-darkgrid")
```

```python
import os
os.chdir("..")
```

```python
df = pd.read_csv("data/raw/semana4.csv")
df["Sex"] = pd.Categorical(df["Sex"])
df.head()
```

```python
model = bmb.Model("Y ~ Income + Sex", data=df)
idata = model.fit(random_seed=787, chains=2)
```

```python
az.summary(idata)
```

```python
az.plot_trace(idata);
```

```python
ax = az.plot_pair(
    idata,
    kind=["scatter", "kde"],
    kde_kwargs={"fill_last": False},
    marginals=True,
    point_estimate="median",
    figsize=(11.5, 5),
)

plt.show()
```
